```python
# data cleaning
df = df.drop_duplicates(keep='first') # removing duplicated rows
df = df.dropna() # removing rows with missing values
#print(df.info()) # verify cleaning
#print(df.isnull().sum()) # verify no missing values
df = df.reset_index(drop=True) # reset index after cleaning
df['Transaction_Amount'] = pd.to_numeric(df['Transaction_Amount'], errors='coerce')
df = df.dropna(subset=['Transaction_Amount']) # drop rows where conversion failed
# fill any remaining NaN values
```

```python
# memory optimization
for col in df.select_dtypes(include=['float64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='float')
for col in df.select_dtypes(include=['int64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='integer')
print(df.memory_usage(deep=True))
df['Transaction_Amount'] = df['Transaction_Amount'].astype('float32')
print(df.Transaction_Amount)
```
