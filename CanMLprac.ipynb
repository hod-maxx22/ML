```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# Topic:simple linear regression with one variable
```

```python
file_path = r'C:/Users/LENOVO YOGA X380/Desktop/ML/canada_per_capita_income.csv'
df = pd.read_csv(file_path)
#df.columns
df.head()
df.info()
```

```python
%matplotlib inline
X = df[['year']]
y = df['per capita income (US$)']
mod = LinearRegression()
mod.fit(X, y)
plt.scatter(df.year, df['per capita income (US$)'], color='blue')
plt.plot(df.year, mod.predict(df[['year']]), color='red')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Year vs Per Capita Income in Canada')
```

```python
X = df[['year']]
y = df['per capita income (US$)']
mod = LinearRegression()
mod.fit(X, y)
pred = mod.predict(X) # predict X column values
print(mod.coef_)  # slope
print(mod.intercept_)  # intercept
print(pred)
```

```python
mod.score(X, y)  # R squared value
```

```python
predict_income_2020 = mod.predict([[2020]])
print(predict_income_2020)
```

```python
%matplotlib inline
#plt.scatter(df.year, df['per capita income (US$)'], color='blue')
#plt.plot(df.year, pred, color='red')
plt.hist(df['per capita income (US$)'],bins=10, color='red', edgecolor='black')
plt.xlabel('Income (US$)')
plt.ylabel('Income Frequency')
plt.title('Income Distribution in Canada')
```

```python
import pickle
# serializing and saving the model to a bin file for future use

with open('mod_pickle','wb') as file_obj:
    pickle.dump(mod,file_obj) # creates a pickle file in binary

with open('mod_pickle','rb') as f:
    model1 = pickle.load(f) # loads your trained model
```
